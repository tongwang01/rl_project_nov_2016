### Figure out how to interpret the action and observation space of CartPole-v0

Make 10 continous steps with action 0. See how observations change.

```python
env = gym.make('CartPole-v0')
observation = env.reset()
print("First observation: ", observation)
for t in range(10):
    env.render()
    action = 0
    observation, reward, done, info = env.step(action)
    print("observation: ", observation)
    print("reward: ", reward)
    print("done: ", done)
    print("info: ", info)
    if done:
        print("Episode finished after {} timesteps".format(t+1))
        break
```

A combination of experimenation and reading the source code. It seems:
* observation = [x, x_dot, theta, theta_dot]
* x is the horizontal position of the cart (right is positive).
* x_dot is the spead of x.
* theta is the angle between the pole and horizontal (clockwise is positive)
